```python
from ultralytics import YOLO

from os import listdir
```

```python
src_dirr = './datasets/v2/train/images/'
imagenes =  list(map( lambda x : './datasets/v2/train/images/' + x, listdir(src_dirr)))
imagenes = list(filter(lambda x : '120' in x, imagenes))
len(imagenes)
```

```python

model = YOLO("./resultados_ejecucciones/yolov9c - PRENTRANADO - batch 16 - 100 epicas - no freeze/weights/best.pt")
model = model.to('cuda')

model.predict(imagenes, save=True, imgsz=320, conf=0.3)

```
