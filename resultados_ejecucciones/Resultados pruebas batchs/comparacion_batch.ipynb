```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d_batch1 = pd.read_csv('./batch 1/results.csv', index_col=0)
d_batch8 = pd.read_csv('./batch 8/results.csv', index_col=0)
d_batch16 = pd.read_csv('./batch 16/results.csv', index_col=0)
d_batch16_f = pd.read_csv('./batch 16 freeze/results.csv', index_col=0)
```

```python
_, axes = plt.subplots(1,2) 

df_maps50 = pd.DataFrame()

df_maps50['Lote tamaño 1'] = d_batch1['       metrics/mAP50(B)']
df_maps50['Lote tamaño 8'] = d_batch8['       metrics/mAP50(B)']
df_maps50['Lote tamaño 16'] = d_batch16['       metrics/mAP50(B)']
#df_maps50['Lote tamaño 16 con congelamiento'] = d_batch16_f['       metrics/mAP50(B)']


df_maps50.plot( title= 'mAP50 | Rendimiento por Lotes', ylabel="mAP50", figsize=(20,10), ax=axes[0])


df_maps50_90 = pd.DataFrame()

df_maps50_90['Lote tamaño 1'] = d_batch1['    metrics/mAP50-95(B)']
df_maps50_90['Lote tamaño 8'] = d_batch8['    metrics/mAP50-95(B)']
df_maps50_90['Lote tamaño 16'] = d_batch16['    metrics/mAP50-95(B)']
#df_maps50_90['Lote tamaño 16 con congelamiento'] = d_batch16_f['    metrics/mAP50-95(B)']

df_maps50_90.plot( title= 'mAP50-95 | Rendimiento por Lotes', ylabel="mAP50-95", figsize=(20,10), ax=axes[1])
```

```python
_, axes = plt.subplots(1,2) 

df_maps50_2 = pd.DataFrame()


df_maps50_2['No Congelamiento'] = d_batch16['       metrics/mAP50(B)']
df_maps50_2['Congelando primeras 10 capas'] = d_batch16_f['       metrics/mAP50(B)']


df_maps50_2.plot( title= 'mAP50 | Rendimiento al aplicar congelamiento', ylabel="mAP50", figsize=(20,10), ax=axes[0])


df_maps50_90_2 = pd.DataFrame()


df_maps50_90_2['Lote tamaño 16'] = d_batch16['    metrics/mAP50-95(B)']
df_maps50_90_2['Lote tamaño 16 con congelamiento'] = d_batch16_f['    metrics/mAP50-95(B)']

df_maps50_90_2.plot( title= 'mAP50-95 | Rendimiento al aplicar congelamiento', ylabel="mAP50-95", figsize=(20,10), ax=axes[1])
```
