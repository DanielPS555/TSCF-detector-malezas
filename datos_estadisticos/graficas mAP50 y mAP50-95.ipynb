```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
from IPython.display import Image, display
```

```python
datos_yolov9c_preentrnado_batch1_no_freeze = pd.read_csv('../resultados_ejecucciones/yolov9 - PREENTRENADO - batch 1 - no  freeze/YOLOv9 - Datasetv2 - PRE ENTRENADO - 100 Ã©pocas - batch1/results.csv', index_col=0)

datos_yolov9c_no_preentrnado_batch1_no_freeze = pd.read_csv('../resultados_ejecucciones/yolov9c - NO PRENTRENADO - batch 1 - 100 epocas - no freeze/results.csv', index_col=0)

datos_yolov9c_preentrnado_batch16_no_freeze = pd.read_csv('../resultados_ejecucciones/yolov9c - PRENTRANADO - batch 16 - 100 epicas - no freeze/results.csv', index_col=0)

datos_yolov9c_preentrnado_batch1_freeze_10 = pd.read_csv('../resultados_ejecucciones/yolov9c - PRENTRENADO - batch 1 - 140 epocas - 10 freeze/grafica en etapas/results etapas 1 - 100.csv', index_col=0)

datos_yolov9c_preentrnado_batch16_freeze_10 = pd.read_csv('../resultados_ejecucciones/yolov9c - PRENTRENADO - batch 16 - 100 epocas - 10 freeze/results.csv', index_col=0)

datos_yolov9c_preentrnado_batch8_no_freeze = pd.read_csv('../resultados_ejecucciones/yolov9c - PRENTRNADO - batch 8 - 100 epocas - no freeze/results.csv', index_col=0)
datos_yolov9c_preentrnado_batch8_no_freeze['metrics/mAP50(B)'] = datos_yolov9c_preentrnado_batch8_no_freeze['metrics/mAP50(B)'].apply(lambda x : x/100)
datos_yolov9c_preentrnado_batch8_no_freeze['metrics/mAP50-95(B)'] = datos_yolov9c_preentrnado_batch8_no_freeze['metrics/mAP50-95(B)'].apply(lambda x : x/100)

datos_yolov9c_preentrnado_batch1_freeze_10_epoca_200 = pd.read_csv('../resultados_ejecucciones/yolov9c - PRENTRENADO - batch 1 - 140 epocas - 10 freeze/grafica en etapas/results etapas 1 - 200.csv', index_col=0)


```

## Pre-entrenado VS o Pre-entrenado

```python


_, axes = plt.subplots(1,2) 

df_maps = pd.DataFrame()

df_maps['Prentrenado'] = datos_yolov9c_preentrnado_batch1_no_freeze['       metrics/mAP50(B)']
df_maps['No Prentrenado'] = datos_yolov9c_no_preentrnado_batch1_no_freeze['       metrics/mAP50(B)']

df_maps.plot( title= 'mAP50 | No Pre-etrenado y Pre entrenado - 100 epocas | batch 1', ylabel="mAP50", figsize=(20,10), ax=axes[0])


df_maps50_90 = pd.DataFrame()

df_maps50_90['Prentrenado'] = datos_yolov9c_preentrnado_batch1_no_freeze['    metrics/mAP50-95(B)']
df_maps50_90['No Prentrenado'] = datos_yolov9c_no_preentrnado_batch1_no_freeze['    metrics/mAP50-95(B)']

df_maps50_90.plot( title= 'mAP50-95 | No Pre-etrenado y Pre entrenado | 100 epocas | batch 1', ylabel="mAP50", figsize=(20,10), ax=axes[1])
```

## Comptacion de Pre entrenamiento y sin Pre entrenamiento aÃ±adiendo Freeze backbone

```python
_, axes = plt.subplots(1,2) 

df_maps = pd.DataFrame()

df_maps['Prentrenado'] = datos_yolov9c_preentrnado_batch1_no_freeze['       metrics/mAP50(B)']
df_maps['No Prentrenado'] = datos_yolov9c_no_preentrnado_batch1_no_freeze['       metrics/mAP50(B)']
df_maps['Prentrenado con Freeze backbone'] = datos_yolov9c_preentrnado_batch1_freeze_10['       metrics/mAP50(B)']

df_maps.plot( title= 'mAP50 | No Pre-etrenado, Pre entrenado (con y sin Freeze backbone) | batch 1', ylabel="mAP50", figsize=(20,10), ax=axes[0])


df_maps50_90 = pd.DataFrame()

df_maps50_90['Prentrenado'] = datos_yolov9c_preentrnado_batch1_no_freeze['    metrics/mAP50-95(B)']
df_maps50_90['No Prentrenado'] = datos_yolov9c_no_preentrnado_batch1_no_freeze['    metrics/mAP50-95(B)']
df_maps50_90['Prentrenado con Freeze backbone'] = datos_yolov9c_preentrnado_batch1_freeze_10['    metrics/mAP50-95(B)']

df_maps50_90.plot( title= 'mAP50-95 | No Pre-etrenado, Pre entrenado (con y sin Freeze backbone) | 100 epocas | batch 1', ylabel="mAP50", figsize=(20,10), ax=axes[1])
```

## Compracion entre resultados con Batch 1, 8 y 16

```python
_, axes = plt.subplots(1,2) 

df_maps = pd.DataFrame()

df_maps['Batch 1'] = datos_yolov9c_preentrnado_batch1_no_freeze['       metrics/mAP50(B)']
df_maps['Batch 8'] = datos_yolov9c_preentrnado_batch8_no_freeze['metrics/mAP50(B)']
df_maps['Batch 16'] = datos_yolov9c_preentrnado_batch16_no_freeze['       metrics/mAP50(B)']

df_maps.plot( title= 'mAP50 |  Pre-entrenado | No Freeze | Batch 1, 8 y 16 | 100 epocas | batch 1', ylabel="mAP50", figsize=(20,10), ax=axes[0])


df_maps50_90 = pd.DataFrame()

df_maps50_90['Batch 1'] = datos_yolov9c_preentrnado_batch1_no_freeze['    metrics/mAP50-95(B)']
df_maps50_90['Batch 8'] = datos_yolov9c_preentrnado_batch8_no_freeze['metrics/mAP50-95(B)']
df_maps50_90['batch 16'] = datos_yolov9c_preentrnado_batch16_no_freeze['    metrics/mAP50-95(B)']

df_maps50_90.plot( title= 'mAP50-95 |  Pre-entrenado | No Freeze | Batch 1, 8 y 16 | 100 epocas | batch 1', ylabel="mAP50", figsize=(20,10), ax=axes[1])
```
